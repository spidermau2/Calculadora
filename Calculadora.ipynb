```python
from IPython.display import clear_output

def validaInt(valor):
    try:
        valor = int(input())
        return valor
    except ValueError:
        print("\nValor inválido, insira novamente...")
        input("Pressione enter para continuar...")
        main()
    
def adicao(valA, valB):
    return valA + valB

def subtracao(valA, valB):
    return valA + valB

def multiplicacao(valA, valB):
    return valA * valB

def divisao(valA, valB):
    return valA // valB

def main():
    opcao = 0
    valA = 0
    valB = 0

    while opcao >= 0 and opcao <= 5:
        clear_output()
        print("1- Adição")
        print("2- Subtração")
        print("3- Multiplicação")
        print("4- Divisão")
        print("5- Sair")
        
        print("Escolha uma operação: ")
        opcao = validaInt(opcao)

        if opcao == 1:
            clear_output()
            print(f"Digite o primeiro valor: ")
            valA = validaInt(valA)
            print(f"Digite o segundo valor: ")
            valB = validaInt(valB)               
            resultado = adicao(valA, valB)
            print(f"\n{valA}+{valB} = {resultado} ")
            input()
        elif opcao == 2:
            clear_output()
            print(f"Digite o primeiro valor: ")
            valA = validaInt(valA)
            print(f"Digite o segundo valor: ")
            valB = validaInt(valB) 
            if valA < 0 or valB >0:
                print("\nFavor inserir negativo no segundo número.")
                input("Enter para continuar...")  
                continue
            resultado = subtracao(valA, valB)
            print(f"\n{valA}{valB} = {resultado} ")
            input()
        elif opcao == 3:
            clear_output()
            print(f"Digite o primeiro valor: ")
            valA = validaInt(valA)
            print(f"Digite o segundo valor: ")
            valB = validaInt(valB)               
            resultado = multiplicacao(valA, valB)
            print(f"\n{valA}*{valB} = {resultado} ")
            input()
        elif opcao == 4:
            clear_output()
            print(f"Digite o primeiro valor: ")
            valA = validaInt(valA)
            print(f"Digite o segundo valor: ")
            valB = validaInt(valB)               
            resultado = divisao(valA, valB)
            print(f"\n{valA} / {valB} = {resultado} ")
            input()
        elif opcao == 5:
            break
        else:
            opcao = 0
            print("\nValor inválido. Favor, insira uma das opções listadas na tela.")
            input("Pressione enter para continuar...")
            
    clear_output()
    print("Programa finalizado!")
    input()
    
main()
        
```
